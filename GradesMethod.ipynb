```python
import numpy as np
```

```python
a = 1
f = lambda x, y: np.tan(x / 2)
alpha = lambda x: np.sin(np.pi * x)
beta1 = lambda t: 0
beta2 = beta1
```

```python
N = 100
T = 1
N2 = 50
tau = T / N2
```

```python
h = 1 / N
xi = lambda i: i * h
tj = lambda j: j * tau
```

```python
ui0 = lambda i: alpha(xi(i))
u0j = lambda j: beta1(tj(j))
unj = lambda j: beta2(tj(j))
```

```python
s = tau * a /  (h ** 2)
fij = lambda i, j: f(xi(i), tj(j))
u = lambda i, j: ui0(i) if j == 0 else \
                u0j(j) if i == 0 else \
                unj(j) if i == N else \
                s * u(i + 1, j - 1) + (1 - 2 * s) * u(i, j - 1) + \
                    s * u(i - 1, j - 1) + tau * fij(i, j - 1)
```

```python
right1 = np.array([u(i, 1) for i in range(N + 1)])
print(right1)
```
